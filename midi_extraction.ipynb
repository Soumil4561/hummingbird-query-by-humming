The All-Mono Method: Combine all channels and keep the highest note
from all simultaneous note events.

```python
#first lets take a midi file
import pretty_midi

location = './test_dataset/archive/Boyzone/Words.mid'

midi_data = pretty_midi.PrettyMIDI(location)

#find number of instruments
print(len(midi_data.instruments))
```

```python
#load each instrument into an array
instruments = []
for instrument in midi_data.instruments:
    instruments.append(instrument)

#find the number of notes in each instrument
for instrument in instruments:
    print(len(instrument.notes))
```

```python
#find length of song
time = midi_data.get_end_time() 
hop = 0.1

#find the number of hops
num_hops = int(time/hop)
print(num_hops)

#how many notes are in each 

#find the highest note in each hop for each instrument
for i in range(num_hops):
    highest_note = 0
    highest_instrument = 0
    for instrument in instruments:
        for note in instrument.notes:
            if note.start < i*hop and note.end > i*hop:
                if note.pitch > highest_note:
                    highest_note = note.pitch
                    highest_instrument = instrument
    print(i, highest_note, highest_instrument)
```
